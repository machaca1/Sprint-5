```python
EXERCICI 2
H0: media de goles de la los equipos italianos es igual que en el resto de paises
H1: media de goles de la los equipos italianos es diferente que en el resto de paises
    
Before conducting the two-sample T-Test we need to find if the given data groups have the same variance. If the ratio of the larger data groups to the small data group is less than 4:1 then we can consider that the given data groups have equal variance.
```

```python
# Import library
import scipy.stats as stats
import numpy as np
 
# Creating data groups


data_group1 = df.loc[df['Country']!='ITA', ['GF']]
data_group2 = df.loc[df['Country']=='ITA', ['GF']]
 
# Print the variance of both data groups
print(np.var(data_group1), np.var(data_group2))
```

```python
# two sample T-test
 
# Import the library
import scipy.stats as stats

# Perform the two sample t-test with equal variances
stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
```

```python
Interpretacion de resultados: 
Como el p-value es mayor que alpha=0.05 no podemos rechazar la hipotesis nula. No hay evidencias suficientes para decir que la media de goles de la liga italiana es diferente que la del resto de equipos europeos.
```

```python
Exercici 3
Is there a statistically significant difference in the mean weight loss among the four diets
¿Hay diferencia significativa entre la media de goles en contra de las ligas de España, Inglaterra y Alemania?
H0: las medias de todas las ligas son iguales
H1: al menos la media de una de las ligas es diferente de las otras
```

```python
# ANOVA
# Example of the Analysis of Variance Test
from scipy.stats import f_oneway
ESP = df.loc[df['Country']=='ESP', ['GF']]
ENG = df.loc[df['Country']=='ENG', ['GF']]
GER = df.loc[df['Country']=='GER', ['GF']]
stat, p = f_oneway(ESP, ENG, GER)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

```python
Interpretacion de resultados:
Como el p-value es mayor que alpha=0.05 no podemos rechazar la hipotesis nula. No hay evidencias suficientes para decir que la media de goles en contra de una de las ligas es diferente a la de las demas.
```
